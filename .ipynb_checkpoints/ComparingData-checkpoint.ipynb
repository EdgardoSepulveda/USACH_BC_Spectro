```python
import numpy as np # Package for scientific computing
import matplotlib.pyplot as plt # Package for plot
%matplotlib notebook
```

Data load (by txt)

```python
# These are the data from Steve's Vernier Sectrometer (1 from here)
sample1 = np.genfromtxt('1ml1Old.txt',skip_header=6)
blank1 = np.genfromtxt('BlankOld.txt',skip_header=6)
# These are the data from Raul's USB 4000 Spectrometer (2 from here) 
sample2 = np.genfromtxt('1ml1New.txt',skip_header=6)
blank2 = np.genfromtxt('BlankNew.txt',skip_header=6)
```

```python
l1_1=sample1[:,0] # wavelength 1
s1_1=sample1[:,1]-0.001 # filter 1 (less the offset)
b1_1=blank1[:,1]-0.001 # blank 1 (less the offset)

l2_1=sample2[:,0] # wavelength 2
s2_1=sample2[:,1]-0.001 # filter 2 (less the offset)
b2_1=blank2[:,1]-0.001 # blank 2 (less the offset)
```

Plot the Intensities Spectras. Figure 1 is the spectra intensity with the Neshyba's Vernier and Figure 2 is the spectra intensity with the Raul's USB 4000 spectrometer

```python
plt.figure()
plt.plot(l1_1,b1_1) # Blank Spectra 1
plt.plot(l1_1,s1_1) # Sample Spectra 1
plt.figure()
plt.plot(l2_1,b2_1) # Blank Spectra 2
plt.plot(l2_1,s2_1) # Sample Spectra 2
```

Select the respective range for each spectra

```python
for i1_i in range(len(l1_1)):
    if l1_1[i1_i]==400: # Select the initial value of the wavelength that you want
        i1=i1_i
for i1_f in range(len(l1_1)):
    if l1_1[i1_f]==650: # Select the final value of the wavelength that you want
        f1=i1_f
subset1_1=[i for i in range(i1,f1)]

for i2_i in range(len(l2_1)):
    if l2_1[i2_i]==400.035057726: # Select the initial value of the wavelength that you want. I extract that value from the respective txt.
        i2=i2_i
for i2_f in range(len(l2_1)):
    if l2_1[i2_f]==650.182557163: # Select the final value of the wavelength that you want
        f2=i2_f
subset2_1=[i for i in range(i2,f2)]

l1_2=l1_1[subset1_1]
s1_2=s1_1[subset1_1]
b1_2=b1_1[subset1_1]

l2_2=l2_1[subset2_1]
s2_2=s2_1[subset2_1]
b2_2=b2_1[subset2_1]
```

Compute and plot the Absorbance

```python
plt.figure()
Ab1_1 = -np.log(s1_2/b1_2) # Absorbance according to Lambert-Beer Law (1)
plt.plot(l1_2,Ab1_1)

plt.figure()
Ab2_1 = -np.log(s2_2/b2_2) # Absorbance according to Lambert-Beer Law (2)
plt.plot(l2_2,Ab2_1)
```

Compute and plot the logarithm of the Absorbance

```python
plt.figure()
l1_3 = np.log(l1_2*0.001) # Logarithm of the wavelenght in micrometers
Ab1_2 = np.log(Ab1_1) # Logarithm of the Absorbance
plt.plot(l1_3,Ab1_2)
plt.figure()
l2_3 = np.log(l2_2*0.001) # Logarithm of the wavelenght in micrometers
Ab2_2 = np.log(Ab2_1) # Logarithm of the Absorbance
plt.plot(l2_3,Ab2_2)
```

Select the portion of the curve that you need

```python
for i3_i in range(len(l1_2)):
    if l1_2[i3_i]==450: # Select the initial value of the wavelength that you want
        i3=i3_i
for i3_f in range(len(l1_2)):
    if l1_2[i3_f]==580: # Select the final value of the wavelength that you want
        f3=i3_f 
subset1_2=[i for i in range(i3,f3)]

plt.figure()
p1 = np.polyfit(l1_3[subset1_2],Ab1_2[subset1_2],1)
b1 = np.exp(p1[1]) # Beta
print ('alpha', -p1[0])
print ('beta', b1)
s1 = np.polyval(p1,l1_3) # Straight
plt.plot(l1_3,Ab1_2)
plt.plot(l1_3,s1,'g')
plt.plot(l1_3[subset1_2],Ab1_2[subset1_2],'o')
plt.grid('on')

for i4_i in range(len(l2_2)):
    if l2_2[i4_i]==450.034782349: # Select the initial value of the wavelength that you want.
        i4=i4_i
for i4_f in range(len(l2_2)):
    if l2_2[i4_f]==580.136883752: # Select the final value of the wavelength that you want
        f4=i4_f
subset2_2=[i for i in range(i4,f4)]

plt.figure()
p2 = np.polyfit(l2_3[subset2_2],Ab2_2[subset2_2],1)
b2 = np.exp(p2[1]) # Beta
print ('alpha', -p2[0])
print ('beta', b2)
s2 = np.polyval(p2,l2_3) # Straight
plt.plot(l2_3,Ab2_2)
plt.plot(l2_3,s2,'g')
plt.plot(l2_3[subset2_2],Ab2_2[subset2_2],'o')
plt.grid('on')
```
